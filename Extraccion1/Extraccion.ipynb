```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Verificamos info del data frame
df.info()
```

```python
#imprimir los primeros 5 registros
df.head()
```

```python
#Filtro por objeto
filtro1=df[df['NOMBRE_CLIENTE']=='PÚBLICO EN GENERAL']
filtro1
```

```python
#Filtro por filas
filtro2=df.iloc[ [42,214,1420] , :]    #Filas no consecutivas
filtro2
```

```python
Filtro3=df.iloc[247:256,: ]           #filas consecutivas
Filtro3
```

```python
#Filtro por columnas
filtro4=df.iloc[: ,2:6 ]           #columnas consecutivas
filtro4
```

```python
filtro5=df.iloc[ : ,[2,5,7]]    #columnas no consecutivas
filtro5
```

```python
#Filtro con indice
df2= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx',index_col=2)
df2
```

```python
#Paso 2: Aplicamos filtro de filas vs columnas
filtro6=df2.loc[["DIEGO ARMANDO LEON DURAN","AARON NUÑO CORDOVA"],["CANT","COSTO_UNITARIO"]]
filtro6
```

```python
#filtro por cabecera
filtro7=df.head(10)
filtro7
```

```python
#filtro por comparacion

filtro8=df[df["PRECIO_UNITARIO"]>20000]
filtro8
```

```python
#filtro Y 
filtro9=df[(df["PRECIO_UNITARIO"] < 9000) & (df["FECHA_DOC"]=="2022-01-14")]
filtro9
```

```python
#Filtro o
filtro10=df[(df["PRECIO_UNITARIO"] < 18000) | (df["FECHA_DOC"]=="2022-01-11")]
filtro10
```

```python
#Filtro not
filtro11=df[~(df["PRECIO_UNITARIO"] < 16000) & ~(df["FECHA_DOC"]=="2022-01-10")]
filtro11
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("filtroObjeto.csv")
filtro2.to_csv("filtrofilas.csv")
filtro4.to_csv("FiltroColumnas.csv")
filtro6.to_csv("FiltroFvsC.csv")
filtro7.to_csv("FiltroCabecera.csv")
filtro8.to_csv("FiltroComp.csv")
filtro9.to_csv("FiltroY.csv")
filtro10.to_csv("FiltroO.csv")
filtro11.to_csv("FiltroNot.csv")
```
